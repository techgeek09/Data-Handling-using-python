```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/upflairs-pvt-ltd/3rd_july_datascience/master/oldbikepriceprediction/Used_Bikes.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['owner'].value_counts()
```

```python
ownership= df['owner'].value_counts().keys()
ownership
```

```python
no_of_bikes = df['bike_name'].value_counts().keys()
no_of_bikes
```

```python
plt.figure(figsize=(10,6))
plt.bar(x='ownership',y='no_of_bikes')
plt.xlabel('owner')
plt.ylabel('no_of_bikes')
plt.title('chart on brands column')
plt.xticks(rotation=60)
plt.show()
```

PIE CHART

```python
country_name
country_population
plt.pie(country_population,labels=country_name)
plt.show() #removes unnecessary information
```

```python
#PIE PLOT
country_name
country_population
plt.pie(country_population,labels=country_name,autopct="%3.3f%%",wedgeprops={'ec':"black",'ls':'dashed','linewidth':4})
plt.show() #removes unnecessary information
```

EXPLOSION- USED TO FOCUS ON A PARTICULAR PART/ SEPARATE IT FROM THE REST

FIGURE- DETERMINES THE SIZE OF THE FIFURE PRODUCED

```python
#PIE PLOT
country_name
country_population
explosion=[0,0,0,0.1]
print(explosion)
plt.figure(figsize=(10,6))
plt.pie(country_population,labels=country_name,autopct="%1.1f%%",wedgeprops={'ec':"black",'ls':'dashed','linewidth':4},explode=explosion)
plt.show() #removes unnecessary information
```
